```python
import json
import jsonlines
import os
import matplotlib.pyplot as plt
import numpy as np
import random
```

```python
def load_data(dataset: list, path: str, label: bool, sort=True):
    fnames = [(path+"/Red/"+i[:2]+"_Red.txt", path+"/Green/"+i[:2]+"_Green.txt",path+"/Blue/"+i[:2]+"_Blue.txt") for i in [i for i in os.walk(path)][1][2]]
    for fname in fnames:
        person = {}
        person["label"] = label
        with open(fname[0],"r") as file:
            person["r"] = [float(i) for i in file.readlines()[1:]]
        with open(fname[1],"r") as file:
            person["g"] = [float(i) for i in file.readlines()[1:]]
        with open(fname[2],"r") as file:
            person["b"] = [float(i) for i in file.readlines()[1:]]
        
        if sort:
            person["r"].sort()
            person["g"].sort()            
            person["b"].sort()
            
        dataset.append(person)
        
def get_filter(dataset: list, color:str):
    if color!="r" and color!="g" and color!="b":
        raise ValueError('Wrong filter: must be "r", "g", "b"')
    new_dataset = []
    for person in dataset:
        new_dataset.append({"label": person["label"], "data": person[color]})
    
    return new_dataset
```

```python
# BC
all_filters = []

load_data(all_filters, "data_orig/Data/BC", True)
load_data(all_filters, "data_orig/Data/Control", False)
```

# Visulaisation of cells fractal dimensions for each person
## Red filter

```python
dataset = get_filter(all_filters, "r")

for i in np.arange(0,len(dataset), 1):
    color="green"
    if dataset[i]["label"]:
        color="red"
    for j in dataset[i]["data"]:
        plt.plot(i,j, "o", markersize=0.5, color=color)
plt.show()
```
